```python
# do all the import stuff
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from PIL import Image
from scipy import signal
import numpy as np
import math
import cv2

# import my modules
import sys
sys.path.append('/home/julian/dev/computer_vision_ws2015/')

import juvis as jv
```

```python
I = cv2.imread('a.jpg')

HSV = cv2.cvtColor(I, cv2.COLOR_BGR2HSV)
H = np.zeros((HSV.shape[0], HSV.shape[1]), dtype=HSV.dtype)
H[:, :] = HSV[:,:,0]

plt.imshow(H)
plt.show()
```

```python
# sobel maybe
Kx = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
Ky = Kx.T

BW = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

Sx = signal.convolve2d(BW, Kx, boundary='symm', mode='same')
Sy = signal.convolve2d(BW, Ky, boundary='symm', mode='same')

Mag = np.sqrt((Sx * Sx) + (Sy * Sy))  # gradient magnitude

jv.paint_mats([H, Sx, Sy, Mag], max_row=2)
```
