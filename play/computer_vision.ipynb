```python
# learning computer vision
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# import my modules
import sys
sys.path.append('/home/julian/dev/computer_vision_ws2015/')

import juvis as jv
# end import my modules

im = np.array(Image.open('smila.png').convert('L'))
im2, cdf = jv.histeq(im)

im3 = jv.compute_average([im, im2])

im4 = im - im2

jv.paint_mats([im, im2, im3, im4])
```

```python
Ma_Face = jv.load_images_as_mat("ma_face", "thb", gray=True)

jv.paint_mats(Ma_Face)

immatrix = jv.vectorize_images(Ma_Face)
```

```python
V, S, immean = jv.pca(immatrix)
M = immean.reshape((40, 40))

modes = [M]
for i in range(7):
    modes.append(V[i].reshape(40, 40))


    
jv.paint_mats(modes)
```
